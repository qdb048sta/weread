```python
import pyautogui
import time
import os
from PIL import Image
import pytesseract
custom_oem_psm_config = r'--oem 3 --psm 3'
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
os.chdir("C:\\Users\\starg\\Pictures\\BlueStacks\\")
time.sleep(5)
file = 'output.txt'
for i in range(0,1530):
    pyautogui.moveTo(3210, 605, duration = 1)
    pyautogui.click(button='left')
    time.sleep(1)
    pyautogui.press("right")
    image_path=None
    time.sleep(0.5)
    for i in os.listdir("C:\\Users\\starg\\Pictures\\BlueStacks\\"):
        if "png" in i:
            image_path=i
            break
    print(image_path)
    image=Image.open(image_path)
    string=pytesseract.image_to_string(image,lang="chi_sim",config=custom_oem_psm_config)
    with open(file, 'a',encoding="utf-8") as f:
        f.write(string)
        f.close()
    os.remove(image_path)
    print(string)
    time.sleep(1)
```
